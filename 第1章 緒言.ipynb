# 1.1 ロボティクスにおける不確実さ
#### ロボティクス : コンピュータで制御された機器を通じて，物理世界を知覚し操作することを扱う科学分野
#### ロボットは物理世界に存在する非常に多くの<u>不確実さ</u>に対処しなければならない


- **ロボット環境**は本質的に予測不可能
    - ハイウェイや個人の家のような環境はよく物が動く
- **センサ**は感知できるものに制限がある
    - 有効範囲や分解能は物理的限界に支配されている
    - 雑音の影響 : 計測値に予測不可能な誤差を与え情報の精度を制限する
    - センサの欠陥検知は困難
- **ロボットの動き** : モータに予測不可能な要素が存在
    - 制御ノイズ，消耗，故障
- **内部モデル** : ソフトウェア中での実世界のモデルには近似が伴う
    - ロボットやロボットの周囲の環境に発生する物理過程の一部を抽象化したものに過ぎない
- **アルゴリズム上の近似**
    - 近似計算により正確さを犠牲に高い実時間応答性

# 1.2 確率ロボティクス
#### 確率ロボティクス : 知覚と行動に関する不確実さを確率論の演算を用いて陽に表現する
(推定(環境情報や行動方策，その未来の予測)を点推定ではなく分布推定する：ベイズ推定)

→ 不確実さが大きい場面では慎重に行動，不確実さを減らすような行動も可能

- **移動ロボットの自己位置推定**
    - ロボットは環境の地図を有し，センサの出力を利用して地図中での自身の位置を求める
    - **大域的自己位置推定** : 既知の環境中のどこかにいロボットが置かれ，そのロボットは全く位置がわからない状態から位置を求めなければならない
    - **信念(belief, $bel$)** : 全ての位置から成る空間中における**ロボットの位置の確率密度関数**
        - (主観確率$bel(x)$ : ロボットが位置$x$にいるであろうと信ずる度合い，頻度論的な確率とは違うことを強調？)
        - Fig1.1(a) 初めはなんら情報がないから(事前分布は)一様分布
        - Fig1.1(b) センサによって<u>ドアのそばにいるという観測$p(z|x)$</u>から「事後」信念が更新
            - (センサにはノイズが入るからドアを観測したからと言ってドアの目の前にいるとは限らないので，ドア付近の確率は分散を持つし，ドアがない場所にも小さい確率が割り振られる)
            - $x$ : 真の位置，$z$ : 観測された位置
            - ドア付近の確率は増加，それ以外は低下
            - ロボット「どこのドアかは知らんけど，どこかのドア**付近**にはいるっしょ！」
        - Fig1.1(c) ロボットの移動後の確率密度分布 : 移動方向へシフトしているが，移動の不確実性により分布はもっと広がる
        - Fig1.1(d) ロボットが再度ドアを観測，(c)での信念とから，自身の位置に関して高い確信を得る
        - (1step前での信念を事前確率，今観測した情報を尤度にして事後確率(今の信念)を更新するイメージ？)
        - Fig1.1(e) ロボットがさらに進んだ後の確率密度分布(信念)
